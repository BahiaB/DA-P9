### Fichier population

```python
#Recuperation des valeurs de population pour 2013 et 2017
pop2013 = pop.loc[pop['Année']==2013,['Zone','Valeur']]
pop2017 = pop.loc[pop['Année']==2017,['Zone','Valeur']]
```

```python
pop = pd.merge(pop2013, pop2017, on='Zone', how='inner')
pop.rename(columns={'Valeur_x' : 'Population 2013', 'Valeur_y' : 'Population 2017'}, inplace=True)
#display( pop[:1])
pop.info()
```

```python
# Calcul de l'evalution de la populatioin entre 2013 et 2017
pop['Pop_evo'] = (pop['Population 2017'] - pop['Population 2013']) / pop['Population 2013']
pop['Population 2017'] = pop['Population 2017'] * 1000
pop.drop(labels='Population 2013', inplace=True, axis=1)
display(pop[:5])
```

```python
# Fusion avec le Data Frame final
complete = pd.merge(left= complete, right= pop, on='Zone', how='inner')

display(complete[:1])
complete.info()
```

### Fichier PIB

```python
# Recuperation des valeur de pib pour 2013 et 2017
pib2013 = pib.loc[pib['Année'] == 2013, ['Zone', 'Valeur']]
pib2017 = pib.loc[pib['Année'] == 2017, ['Zone', 'Valeur']]

pib = pd.merge(pib2013, pib2017, on='Zone', how='inner')
pib.rename(columns={'Valeur_x' : 'pib 2013', 'Valeur_y' : 'pib 2017'}, inplace=True)
```

```python
# Calcul du taux d'evolution du pib
pib['Evolution pib'] = (pib['pib 2017'] - pib['pib 2013']) / pib['pib 2013']
pib = pib[['Zone','pib 2017', 'Evolution pib']]
display(pib[:1])
```

```python
complete = pd.merge(left= complete, right= pib, on='Zone', how='inner')
display(complete[:1])
#complete.info()
```
